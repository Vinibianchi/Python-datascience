```python
import seaborn as sb
import matplotlib.pyplot as plt
import random as rd
%matplotlib inline
```

```python
# Datasets de exemplo do seborn
sb.get_dataset_names()
```

```python
# Carredando data set num dataFrame
titanic = sb.load_dataset('titanic')
```

```python
titanic.head()
```

```python
tips  = sb.load_dataset('tips')
tips.head()
```

```python
fly = sb.load_dataset('flights')
fly.head()
```

```python
# Pivotiando um dataFrame com o método pivot_table
fly = fly.pivot_table(values='passengers',index = 'month', columns='year')
```

```python
fly.head()
```

# distplot
## Montanto Histograma  
### frequencia no eixo y

```python
# tips é o valor da gorjeta
# bins é amplitude do agrupamento da variaével analisada
sb.distplot(tips['tip'], kde = True, bins =10)
```

# Correlacionando Gráficos
## Gráficos Lineares

## Joinplot

### Junta Historamas com plots e é útil para ver a relação de duas variáveis

```python
# Vamos verificar total_bill x tips

sb.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sb.jointplot(x='fare',y='age',data=titanic)
```

# Plotando dados categóricos 

```python
# como temos mutios preços (fare), o gráfico fica ruim. nesse caso faz mais sentido trabalhar com var categóricas
sb.countplot(y='fare',data=titanic)
```

```python
sb.countplot(data=titanic, x='sex')
```

# BoxPlot

```python
sb.boxplot(x='day',y='tip', data=tips)
```

```python
# Segmentando por sexo com HUE
sb.boxplot(x='day',y='tip', data=tips, hue='sex')
```

# grafico de freq violino


```python
sb.violinplot(x='day',y='tip',data=tips)
```

```python
sb.violinplot(x='day',y='tip',data=tips, hue='sex')
```

```python
sb.violinplot(x='day',y='tip',data=tips, hue='sex', split=True)
```

# Stripplot
