# Tratando dados NaN

```python
import pandas as pd
dados = pd.read_csv('Datasets/aluguel_residencial.csv', sep = ';')
dados.head()
```

## Formas de Identificar NaN

```python
# Informa a quantidade de linhas e not null
dados.info()
```

```python
# Data frame Booleano de valores nulos
dados.isnull()
```

```python
# Data frame Booleano de valores not nulos
dados.notnull()
```

## Filtrando um Data Frame de valores `Null e Not Null` em uma coluna


```python
# DF só de NaN em Valor
dados[dados['Valor'].isnull()]
```

```python
# DF só de Not Null em Valor
dados[dados['Valor'].notnull()]
```

```python
# Outra Forma DF só de Not Null em Valor
dados[~dados['Valor'].isnull()]
```

## Filtrando um DF de valores `Null e Not Null` com Condições

```python
dados.head()
```

```python
# Selecionado Tipo apartamento e Condominio null
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
# Quantidade de exclusões de de Null
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```python
# Apartamento e Condominio not null
dados[~selecao].head(15)
```

## Preenchendo Valores NaN Interpolação 

```python
data = [0.5, None, None, 0.52, 0.54, None, None, 0.59, 0.6, None, 0.7]
s = pd.Series(data)
```

```python
# Substituindo NaN por 0
# PS: Para manter a alteração precisa passar inplace=True
dados.fillna(0)
```

```python
# Preenchendo Valores Na com Dicionário de Duas (ou mais) colunas.
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```
